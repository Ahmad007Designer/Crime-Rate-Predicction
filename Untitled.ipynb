```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.neighbors import KNeighborsRegressor
from sklearn import tree
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn import metrics
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming 'dataset' is a pandas DataFrame with appropriate columns
fig, ax = plt.subplots(2, 3, figsize=(18, 10), sharey=True)  # Removed log=True

colors = ['blue', 'green', 'red', 'purple', 'orange', 'pink', 'cyan', 'brown', 'gray', 'olive', 'navy']

# Loop through the 6 subplots (2 rows x 3 columns)
for i in range(6):
    row = i // 3 
    col = i % 3   
    ax[row, col].bar(dataset['City'], dataset[dataset.columns[i + 2]], color=colors[i])
    ax[row, col].set_title(f'City vs {dataset.columns[i + 2]}')
    ax[row, col].tick_params(axis='x', rotation=90)

    # Set log scale for the y-axis
    ax[row, col].set_yscale('log')

plt.tight_layout()
plt.show()

```

```python
new_df = pd.DataFrame(columns=['Year', 'City', 'Population (in Lakhs) (2011)+', 'Number Of Cases', 'Type'])
for i in range(3, 13):
    temp = dataset[['Year', 'City', 'Population (in Lakhs) (2011)+']].copy()
    temp['Number Of Cases'] = dataset[[dataset.columns[i]]]
    temp['Type'] = dataset.columns[i]
    
    new_df = pd.concat([new_df, temp])
```

```python
new_df
```

```python
new_df['Crime Rate'] = new_df['Number Of Cases'] / new_df['Population (in Lakhs) (2011)+']
```

```python
new_df = new_df.drop(['Number Of Cases'], axis=1)
```

```python
new_dataset = pd.read_excel("new_dataset.xlsx", sheet_name="Sheet1")
```

```python
le = LabelEncoder()
```

```python
new_dataset['City'] = le.fit_transform(new_dataset['City'])
mapping = dict(zip(le.classes_, range(len(le.classes_))))
```

```python

# Saving the mapping file for further use
file = open('City_Mapping.txt', 'wt')
for key,val in mapping.items():
    print(str(key) + " - " + str(val) + '\n')
    file.write(str(key) + " - " + str(val) + '\n')
```

```python
file = open('City_Mapping.txt', 'wt')
for key,val in mapping.items():
    print(str(key) + " - " + str(val) + '\n')
    file.write(str(key) + " - " + str(val) + '\n')
```

```python

new_dataset['Type'] = le.fit_transform(new_dataset['Type'])
mapping = dict(zip(le.classes_, range(len(le.classes_))))
```

```python
file = open('Type_Mapping.txt', 'wt')
for key,val in mapping.items():
    print(str(key) + " - " + str(val) + '\n')
    file.write(str(key) + " - " + str(val) + '\n')
```

```python
new_dataset
```

```python

x = new_dataset[new_dataset.columns[0:4]].values
x
```

```python

y = new_dataset['Crime Rate'].values
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=50)
```

```python

x_train
```

```python
y_train
```

```python
model1 = svm.SVR()
model1.fit(x_train, y_train)
y_pred = model1.predict(x_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('R2 score:', metrics.r2_score(y_test, y_pred))
```

```python
model2 = KNeighborsRegressor(n_neighbors=2)
model2.fit(x_train, y_train)
y_pred = model2.predict(x_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('R2 score:', metrics.r2_score(y_test, y_pred))
```

```python
model3 = tree.DecisionTreeRegressor()
model3.fit(x_train, y_train)
y_pred = model3.predict(x_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('R2 score:', metrics.r2_score(y_test, y_pred))
```

```python
model4 = RandomForestRegressor(random_state=0)
model4.fit(x_train, y_train)
y_pred = model4.predict(x_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('R2 score:', metrics.r2_score(y_test, y_pred))
```

```python
import numpy as np
import matplotlib.pyplot as plt

algorithms = ['Decision Tree', 'Random Forest', 'XGBoost']
mae = [2.8914, 2.4879, 3.0256]
mse = [34.9666, 21.3629, 25.2270]
r2_score = [0.8891, 0.9322, 0.9200] 
sorted_indices = np.argsort(r2_score)[::-1]
algorithms = [algorithms[i] for i in sorted_indices]
mae = [mae[i] for i in sorted_indices]
mse = [mse[i] for i in sorted_indices]
r2_score = [r2_score[i] for i in sorted_indices]
# Plotting
bar_width = 0.4
x = np.arange(len(algorithms))
fig, ax = plt.subplots(1, 3, figsize=(18, 5))

ax[0].bar(x, mae, color='blue', width=bar_width)
ax[0].set_xticks(x)
ax[0].set_xticklabels(algorithms, rotation=15, ha="right")
ax[0].set_title('Mean Absolute Error')
ax[0].set_ylabel('MAE')

ax[1].bar(x, mse, color='green', width=bar_width)
ax[1].set_xticks(x)
ax[1].set_xticklabels(algorithms, rotation=15, ha="right")
ax[1].set_title('Mean Squared Error')
ax[1].set_ylabel('MSE')

ax[2].bar(x, r2_score, color='red', width=bar_width)
ax[2].set_xticks(x)
ax[2].set_xticklabels(algorithms, rotation=15, ha="right")
ax[2].set_title('RÂ² Score (Higher is Better)')
ax[2].set_ylabel('RÂ² Score')

plt.tight_layout()
plt.show()

```
